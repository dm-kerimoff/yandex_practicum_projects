В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**ОПИСАНИЕ ДАННЫХ**

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры

- EU_sales — продажи в Европе (миллионы проданных копий)

- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

### Обработка пропусков

```python
games.isna().sum()
```

```python
print(f"Пропущенные строки составляют {games['year_of_release'].isna().mean():.2%} от общего количества строк")
```

Поэтому эти строки можно удалить

```python
# удалим строки с пропусками
games = games.dropna(subset=['year_of_release']).reset_index(drop=True)
```

```python
games.isna().sum()
```

**Пропуски в столбце `rating`**

```python
games.isna().sum()
```

Поскольку пропущенных строк в столбцах `critic_score`, `user_score`, `rating` критически много, то просто так удалять строчки нельзя - грозит искажением данных и последующего анализа(GIGO - Garbage In Garbage Out), то с каждым столбцом необходимо работать отдельно.

Мы не можем заменить пропуски медианным или средним значением, т.к это критически исказит данные и их результаты. Т.к переменные `critic_score` и `user_score` количественные,то мы не можем заменить пропуски заглушкой `UNK` от слова 'Unknown', поскольку тип данных изменится с `float` на `object`. Поэтому строки с пропусками в данных столбцах стоит оставить в этих столбцах

В столбце `rating` также очень много пропусков, оставим их из-за отсутствия возможности заменить пропуски

**Ранее было отмечено, что в столбце `user_score` присутствует значение `tbd`. Посмотрим, сколько таких значений**

```python
len(games.query('user_score == "tbd"'))
```

Весьма много значений, чтобы удалить столько строк или заменить медианным значением, поэтому заменим эти значения на `Nan`

```python
games['user_score'] = games['user_score'].replace('tbd', np.nan)
```

**Выводы**
 - Удалены 2 строки с пропусками в столбце `name` и `genre` и 269 строк с пропусками в столбце `year_of_release`
 - Пропуски в столбцах `critic_score` и `user_score` мы оставили
 - Пропуски в столбце `rating` мы заменили на значение `UNK`


 - **Пропуски в столбце `rating` могут быть вызваны следующими причинами:**
   1. Различия в оценке игры ассоциациями разных государств. Entertainment Software Rating Board - американская ассоциация, а значит, игры, которые издаются в других государствах, могут оцениваться по собственным системам оценок. Поэтому, из-за различий в этих оценках значение rating могли оставлять пустыми.
   2. Игра могла просто не получить рейтинг по техническим причинам со стороны ESRB
   3. Не исключено, что возможны случаи, когда однозначно определить рейтинг игры невозможно
   
   
 - **Пропуски в столбце `critic_score` и `user_score` могли быть вызваны тем, что** было недостаточно отзывов от пользователей, чтобы оценить игры

### Влияние отзывов и критики на продажи у популярных платформ

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграммы рассеяния и посчитаем корреляцию между отзывами и продажами. Сформулируем выводы.

**Построим диаграммы рассеяния для каждой платформы и посчитаем коэффициент корреляции Пирсона для каждой платформы**

```python
# создадим функцию для построения диаграмм рассеяния и подсчёта коэффициента корреляции
def scatter_corr(platform):
    fig, axs = plt.subplots(1,2, figsize=(10,5))
    games_modern.query('platform == @platform').plot(kind='scatter', x='user_score', y='total_sales', ax=axs[0], grid=True)
    games_modern.query('platform == @platform').plot(kind='scatter', x='critic_score', y='total_sales', ax=axs[1], grid=True)
    fig.suptitle(platform, fontsize=13)
    axs[0].set(xlabel='Оценка пользователей', ylabel='Общие продажи')
    axs[1].set(xlabel='Оценка критиков', ylabel='Общие продажи')
    axs[0].xaxis.get_label().set_fontsize(11)
    axs[0].yaxis.get_label().set_fontsize(11)
    axs[1].xaxis.get_label().set_fontsize(11)
    axs[1].yaxis.get_label().set_fontsize(11)
    plt.show()
    user_corr = games_modern.query('platform == @platform')['total_sales'].corr(games_modern.query('platform == @platform')['user_score'])
    critic_corr = games_modern.query('platform == @platform')['total_sales'].corr(games_modern.query('platform == @platform')['critic_score'])
    print(f'Коэффициент корреляции между оценками пользователей и продажами на платформе {platform} равен {user_corr}')
    print(f'Коэффициент корреляции между оценками критиков и продажами на платформе {platform} равен {critic_corr}')
```

```python
for platform in top6_platforms['platform'].unique():
    scatter_corr(platform)
```

**Выводы**

- По диаграммам рассеяния и значениям корреляции можно отметить, что почти для каждой платформы между объемами продаж и оценками пользователей и критико присутсвует зависимость, но она слишком мала.
- Однако также можно отметить, что связь между объёмами продаж и оценками критиков сильнее, чем оценок пользователей: пользователи охотнее прислушиваются к мнениям экспертов

### Влияние рейтинга ESRB на продажи в отдельном регионе

```python
# создадим столбец с описанием рейтинга ESRB
def ESRB_describe(rating):
    if rating == 'M':
        return '17+'
    elif rating == 'E':
        return 'Для всех'
    elif rating == 'E10+':
        return '10+'
    elif rating == 'T':
        return '13+'
    elif rating == 'EC':
        return '6+'
    elif rating == 'RP':
        return 'Не определён'
    else:
        return 'Не известен'
```

```python
games_modern['ESRB_describe'] = games_modern['rating'].apply(ESRB_describe)
```

## Проверка гипотез

### Гипотеза №1:

Проверим гипотезу: cредние пользовательские рейтинги платформ Xbox One и PC одинаковые;

**H0: Средний рейтинг пользователей (`user_score`) на платформе Xbox One = (равен) среднему рейтингу пользователей (`user_score`) на платформе PC**

**H1: Средний рейтинг пользователей (`user_score`) на платформе Xbox One <>(не равен) среднему рейтингу пользователей (`user_score`) на платформе PC**

**alpha = 0.05**

**Перед использовванием t-теста о равенстве средних двух генеральных совокупностей проверим размеры выборок:**

```python
x_one = games_modern.query('platform == "XOne"')['user_score'].dropna()
pc = games_modern.query('platform == "PC"')['user_score'].dropna()
print(f'Размеры выборок {len(pc)} и {len(x_one)} соответственно')
```

Размеры выборок существенно отличаются. Это нужно учесть при настройках теста

```python
alpha = 0.05
results = st.ttest_ind(x_one, pc, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**

Мы не можем отвергнуть нулевую гипотезу. Следовательно, действительно, есть смысл, согласно нулевой гипотезе, считать равным средний рейтинг пользователей на обоих платформах XOne и PC.


**Также выведем средние значения данных выборок**

```python
x_one.mean()
```

```python
pc.mean()
```

### Гипотеза №2

Проверим другую гипозету: "Средние пользовательские рейтинги жанров `Action` и `Sports`разные"

**H0: Средний рейтинг пользователей (`user_score`) жанра `Action` = (равен) среднему рейтингу пользователей (`user_score`) жанра `Sports`**

**H1: Средний рейтинг пользователей (`user_score`) жанра `Action` <> (не равен) среднему рейтингу пользователей (`user_score`) жанра `Sports`***

**alpha = 0.05**

**Перед использовванием t-теста о равенстве средних двух генеральных совокупностей проверим размеры выборок:**

```python
action = games_modern.query('genre=="Action"')['user_score'].dropna()
sports = games_modern.query('genre=="Sports"')['user_score'].dropna()
print(f'Размеры выборок {len(action)} и {len(sports)} соответственно')
```

**Размеры выборок также существенно отличаются: учтём это при статистическом тесте**

```python
alpha = 0.05
results = st.ttest_ind(action,sports, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
action.mean()
```

```python
sports.mean()
```

**Вывод**

P-value оказалось исчезающе малым, поэтому нулевая гипотеза нами отверглась. А значит мы можем с большей уверенностью рассуждать о том, **средние пользовательские рейтинги жанров Action и Sports различаются**.

***ПРОМЕЖУТОЧНЫЕ ВЫВОДЫ***

- В данных задачах мы тестируем гипотезы о равенстве средних двух генеральных совокупностей

В таком случае нулевая гипотеза формулируется следующим образом: "Средние значения генеральных совокупностей равны"
а) Средняя оценка пользователей на одной платформе **равна** средней оценке пользователей на другой платформе
б) Средняя оценка пользователей одного жанра **равна** средней оценке пользователей другого жанра

- Альтернативная гипотеза при этом может быть односторонней(левостороннй или правосторонней) или двусторонней.

В данных задачах альтернативная гипотеза является двусторонней, т.е
а) Средняя оценка пользователей на одной платформе **НЕ равна** средней оценке пользователей на другой платформе
б) Средняя оценка пользователей одного жанра **НЕ равна** средней оценке пользователей другого жанра

При тестировании гипотез применяется t-критерий, который проверяет, существенно ли отличаются средние генеральных совокупностей.
Для применения данного теста должны выполнять следующие условия:
- Наблюдения в выборке должны быть независимы друг от друга
- Наблюдения в каждой выборке распределяются нормальным образом
- Наблюдения в каждой выборке имеют одинаковую дисперсию
