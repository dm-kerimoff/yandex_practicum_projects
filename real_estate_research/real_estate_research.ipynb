# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

## Предобработка данных

**Для удобства получения общей информации о столбцах напишем функцию, которую будем неоднократно использовать при получении информации**

```python
def describe_columns(col):
    print(f'Тип данных столбца: {data[col].dtypes}')
    print(f'Общее количество пропусков в столбце: {data[col].isna().sum()}', '\n')
    print(data[col].describe())
```

**Для удобства заполнения пропусков в столбцах по группам значений другого столбца напишем функцию, которую также будем неоднократно использовать**

```python
def fill_skip_median(col, category):
    for val in data[category].unique():
        data.loc[(data[col].isna()) & (data[category] == val), col] = data[data[category] == val][col].median()
```

Поработаем над аномальными значениями основных столбцов, которые влияют на стоимость квартиры (цена, площади, высота потолков, количество комнат, этаж). В рамках проекта воспользуемся либо методом межквартильного размаха IQR, либо по собственным логичным рассуждениям

```python
# функция реализует избавление от выбросов в столбце методом межквартильного размаха
def blowouts_iqr(col):
    q75, q25 = np.percentile(data[col], [75 ,25])
    iqr = q75 - q25
    return data.loc[(data[col] >= (q25 - 1.5*iqr)) & (data[col] <= (q75 + 1.5*iqr))]
```

```python
# функция реализует поиск границ выбросов столбца с помощью межквартильного размаха IQR
def blowouts_info(col):
    q75, q25 = np.percentile(data[col], [75 ,25])
    iqr = q75 - q25
    #посчитаем допустимые границы нормальных значений с помощью межквартильного размаха:
    blowouts_min = q25 - 1.5*iqr
    blowouts_max = q75 + 1.5*iqr
    # выбросы столбца по значениям blowouts_min blowouts_max:
    print(f'Нижняя граница выбросов столбца {col}', blowouts_min)
    print(f'Верхняя граница выбросов столбца {col}', blowouts_max)
```
